## Creating a formula to rank these 100 locations. Explaining why did I choose this ranking/formula, formula and the final rankings.

#### One of the soltion to this problem is creating a Ranked list of different Locations as per category; lets say a person wants to buy a car from a location having best administration or best overall experience then the person can just refer to the ranked list as per category and choose the location having best rating for administration or best overall experience.

#### Approach: 1) We can make use of 3 KPI's to derive a ranked list namely no of reviews, average rating and net sentiment
####           2) Adding the 3 KPI's to derive a Net Value column, we just cant assign a rank to the locations based on Net Value
####           3) As Net Value is just a distribution we need to know how the locations can be organized such that the distribution makes sense. For that we need to calculate z-score for every location such that it will give us an idea how locations are distributed in this distribution and how far are the locations from the mean of Net Value.
####           4) Formula for Z-Score = [Net value - Mean(Net Value)] / standard deviation(Net Value)
####           5) The Z-Score of the Mean of Net value is 0 which means any location having a higher Z-Score has a better Net Value which in turns mean that location will have a higher rank
####           6) Once the z-score is calculated, the rankings are assigned to the locations within the categories by arranging the z-score in descending order within the categories. The location with highest z-score in a category will have Rank #1 and so on and so forth

```python
rev_catg_jn = reviews.join(categories.set_index("review_id"), on='review_id', how='left')
```

```python
cmbd = (rev_catg_jn.groupby(["category","location_id"])
           .agg({'review_id':'count', 'rating': 'mean', 'sentiment': 'mean'})
           .rename(columns={'review_id':'no of reviews', 'rating': 'avg rating', 'sentiment': 'net sentiment'})
 )
```

```python
x = cmbd["no of reviews"]
y = cmbd["avg rating"]
z = cmbd["net sentiment"]
res = x + y + z
```

```python
cmbd["net value"] = res
```

```python
cmbd
```

```python
std_dev = cmbd["net value"].values.std(ddof=1)
```

```python
res = (cmbd["net value"] - cmbd["net value"].mean())/std_dev
```

```python
cmbd["z-score"] = res
```

```python
cmbd
```

```python
rank = cmbd.groupby(level=0, group_keys=False)
```

```python
rank.get_group(cmbd.index.levels[0][:1].values[0])
```

```python
ranked_list = pd.DataFrame()
ranked_grp = pd.DataFrame()
for i in range(1, len(cmbd.index.levels[0])+1):
    ranked_grp = rank.get_group(cmbd.index.levels[0][:i].values[i-1])
    ranked_grp = ranked_grp.sort_values(by=["z-score"],ascending=False)
    ranked_grp["rank"] = range(len(ranked_grp))
    ranked_grp["rank"] += 1
    ranked_list = ranked_list.append(ranked_grp)
        
```

## Finding at least one location that is mis-ranked by the above algorithm (i.e. they should probably be higher or lower). Why are they mis-ranked? How to tackle this

#### The major problem with the above algorithm is that locations having bad average rating and net sentiment value has higher rank just becuase it has  more no of reviews. This is logically not correct thus giving birth to the problem of over-rating and under-rating.

#### The location 4962_205 is mis-ranked by the above algorithm as it is evident that is has a mere average rating of 1.79 and net sentiment value of 0.5 but just becasue it has more no of reviews and a higher z-score it has been ranked higher.

#### Solution to the problem of Under-Rating and Over-Rating:
#### 1) Instead of adding all the 3 KPI's we will add the 2 KPI's i.e. average rating and net sentiment to derive a Net Value column but we cant just assign a rank to the locations based on Net Value
#### 2) As Net Value is just a distribution we need to know how the locations can be organised such that the distribution makes sense. For that we need to calculate z-score for every location such that it will give us an idea how locations are distributed in this distribution and how far are the locations from the mean of Net Value.
#### 3) Formula for Z-Score = [Net value - Mean(Net Value)] / standard deviation(Net Value)
#### 4) The Z-Score of the Mean of Net value is 0 which means any location having a higher Z-Score has a better Net Value which in turns mean that location will have a higher rank. But locations having same z-score will same rank, this challenge can be further solved by introducing no of reviews. If multiple locations have same z-score then higher the no of reviews higher will be rank of that location in the ranked list

```python
cmbd3 = (rev_catg_jn.groupby(["category","location_id"])
           .agg({'rating': 'mean', 'sentiment': 'mean'})
           .rename(columns={'rating': 'avg rating', 'sentiment': 'net sentiment'})
 )
```

```python
rate = cmbd3["avg rating"]
sent = cmbd3["net sentiment"]
```

```python
net_value = rate + sent
```

```python
cmbd3["net value"] = net_value
```

```python
std_dev3 = cmbd3["net value"].values.std(ddof=1)
z_scr = (cmbd3["net value"] - cmbd3["net value"].mean())/std_dev3
cmbd3["z-score"] = z_scr
cmbd3
```

```python
rank3 = cmbd3.groupby(level=0, group_keys=False)
```

```python
ranked_list3 = pd.DataFrame()
ranked_grp3 = pd.DataFrame()
for i in range(1, len(cmbd3.index.levels[0])+1):
    ranked_grp3 = rank3.get_group(cmbd3.index.levels[0][:i].values[i-1])
    ranked_grp3 = ranked_grp3.sort_values(by=["z-score"],ascending=False)
    ranked_grp3["rank"] = range(len(ranked_grp3))
    ranked_grp3["rank"] += 1
    ranked_list3 = ranked_list3.append(ranked_grp3)
```

```python
ranked_list3
```

###### Location's with same z-score should have same ranking 

```python
final_df = pd.DataFrame()
for i in range(1, len(cmbd3.index.levels[0])+1):
    df = rank3.get_group(cmbd3.index.levels[0][:i].values[i-1])
    df = df.sort_values(by=["z-score"],ascending=False)
    lst = df["z-score"]
    lst = list(lst)
    lst1 = []
    r = 1
    lst1.append(r)
    for l in range(1, len(lst)):
        if lst[l-1] == lst[l]:
            lst1.append(r)
        else:
            r = r+1
            lst1.append(r)
    cat = df.index.get_level_values('category')
    lst1 = pd.DataFrame(lst1)
    lst1.set_index(cat)
    df["rank"] = lst1.values
    final_df = final_df.append(df)
```

```python
final_df
```

###### Here the problem of under-rating and over-rating has been solved but the major concern of a buyer still remain; how to choose a location?. This can be solved by introducing the no of reviews. If 2 places have same ranking for same category then no of reviews might help to decide which place is a better choice.

```python
cmbd4 = (rev_catg_jn.groupby(["category","location_id"])
           .agg({'review_id':'count', 'rating': 'mean', 'sentiment': 'mean'})
           .rename(columns={'review_id':'no of reviews', 'rating': 'avg rating', 'sentiment': 'net sentiment'})
 )
```

```python
rate1 = cmbd4["avg rating"]
sent1 = cmbd4["net sentiment"]
net_value1 = rate1 + sent1
cmbd4["net value"] = net_value1
```

```python
std_dev4 = cmbd4["net value"].values.std(ddof=1)
z_scr1 = (cmbd4["net value"] - cmbd4["net value"].mean())/std_dev4
cmbd4["z-score"] = z_scr1
```

```python
rank4 = cmbd4.groupby(level=0, group_keys=False)
```

```python
final_df2 = pd.DataFrame()
for i in range(1, len(cmbd4.index.levels[0])+1):
    df = rank4.get_group(cmbd4.index.levels[0][:i].values[i-1])
    df = df.sort_values(by=["z-score","no of reviews"],ascending=[False,False])
    x1 = f[["z-score","no of reviews"]]
    x1 = list(x1.values)
    lst = []
    r = 1
    lst.append(r)
    for x in range(1, len(x1)):
        res1 = x1[x-1]
        res2 = x1[x]
        if res1[0] == res2[0]:
            if res1[1] == res2[1]:
                lst.append(r)
            else:
                r = r+1
                lst.append(r)
        else:
            r = r+1
            lst.append(r)
    cat = f.index.get_level_values('category')
    c = pd.DataFrame(lst)    
    c.set_index(cat)
    f["rank"] = c.values
    final_df2 = final_df2.append(f)
```

###### The final ranking explains that the locations are ranked as per the z-score but if 2 locations have same z-score then the no of reviews can be considered. In this case a location having higher no of reviews value will have a higher rank. But if no of reviews are also same then we need some other metrics; such as proximity of the location to the buyer's place of residence or no of years the auto-dealership has been operating etc.; to decide which location is better and have a higher rank than than the location having same z-score and no of reviews.
